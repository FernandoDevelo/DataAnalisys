```python
## Temos colunas com texto, é preciso conveter com get dummies
```

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

## Vamos dividir os dados em característica e alvo ( features e target )

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
## Vamos usar o decision tree ou árvore de decisão para classificar
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
## Predições e avaliação do modelo
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

## Os scores de precision, recall e f1 não estão muito bons, decision tree normalmente não tem um bom resultado se não usada
##com random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

```python
## Vamos fazer a predição e avaliação do random forest agora..
```

```python
predictions = rfc.predict(X_test)
```

**Avaliação do modelo

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
